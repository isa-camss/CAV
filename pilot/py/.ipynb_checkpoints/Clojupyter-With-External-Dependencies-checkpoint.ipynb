<h1><b>Adding external dependencies to the Clojure Jupyter Notebook</b></h1>

```clojure
(require '[clojupyter.misc.helper :as helper])
(helper/add-dependencies '[incanter "1.5.7"])
(use '(incanter core stats charts io))
```
